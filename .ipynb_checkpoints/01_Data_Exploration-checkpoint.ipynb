## Remove columns with constant values

```python
constant_columns = []
for col in df.columns:
    if df[col].nunique() == 1:
        constant_columns.append(col)
print(f'columns with constant values are {constant_columns}')
```

```python
df_clean = df.drop(columns=constant_columns)
print(f'{constant_columns} dropped')
```

## Employee number is an index column

```python
df_clean.index = df_clean.EmployeeNumber 
df_clean = df_clean.drop(columns=['EmployeeNumber'])
```

## Check each column and its importance towards the target variable

```python
import matplotlib.pyplot as plt
from IPython.display import HTML, display
def draw_data(d:pd.DataFrame, col_name):
    
    print(f'\ncolname {col_name}')
    d[col_name].hist()
    plt.xlabel(col_name)
    plt.show()
    
    a =utils.make_columns_numerical(d[['Attrition',col_name]])
    a.pivot(columns=['Attrition'],  values = col_name).plot.hist()
    plt.xlabel(col_name)
    plt.show()
    

    perc = utils.make_percentage(d, col_name)
    display(HTML(perc.to_html()))
  
```

```python
col_name = 'BusinessTravel'
draw_data(df_clean, col_name)
```

```python
all_cols = list(df_clean.columns)
all_cols.remove('Attrition')
for col in all_cols:
    draw_data(df_clean, col)
```
